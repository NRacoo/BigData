```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
```

```python
x, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)

plt.scatter(x[:, 0], x[:,1], s=30)
plt.title("Datasets awal (tanpa label)") 
plt.show()
```

```python
inertia = []
K = range(1, 10)

for k in K: 
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(x)
    inertia.append(kmeans.inertia_)
    
plt.plot(K, inertia, 'bo-')
plt.xlabel('Jumlah Kluster(k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
y_kmeans = kmeans.fit_predict(x)

plt.scatter(x[:, 0], x[:, 1], c=y_kmeans, s=30, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker="X", label="Centroid")
plt.title("Hasi Clustering K-Means (k=4)")
plt.legend()
plt.show()
```

# Moon Data

```python
X, y = make_moons(n_samples=300, noise=0.06, random_state=0)
X_scaled = StandardScaler().fit_transform(X)

plt.scatter(X_scaled[:, 0], X_scaled[:,1], s=30)
plt.title("Datasets awal (tanpa label)") 
plt.show()
```

```python
dbscans = DBSCAN(eps=0.3, min_samples=4)
y_dbscans = dbscans.fit_predict(X_scaled)


plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y_dbscans, cmap='Paired', s=30)
plt.title("Hasil clustering DBSCAN")
plt.show()

```
